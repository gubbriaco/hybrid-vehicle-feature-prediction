```python
import os
import pandas as pd
import numpy as np
```

```python
data_dir = './data'
measurement_data_dir = f'{data_dir}/measurement'
os.listdir(measurement_data_dir)
```

```python
measurement_data_overview_path = os.path.join(measurement_data_dir, 'Overview.xlsx')
measurement_data_overview = pd.read_excel(measurement_data_overview_path)
measurement_data_overview
```

```python
trips_dir = os.path.join(measurement_data_dir, 'trips')
os.listdir(trips_dir)
```

```python
trips = []
csv_list = [f for f in os.listdir(trips_dir) if f.endswith('.csv')]
for i in range(len(csv_list)):
    trip_name = csv_list[i]
    trip_path = os.path.join(trips_dir, trip_name)
    trip = pd.read_csv(trip_path)
    trips.append(trip)
```

Salvo le colonne corrispondenti ad ogni dataframe corrispondente ad ogni file .csv presente nella directory

```python
cols_trips = []
for i in range(len(trips)):
    cols_trips.append(trips[i].columns)
cols_trips
```

Calcolo il numero di colonne presenti nel dataframe corrispondente ad ogni file .csv presente nella directory

```python
no_cols_trips = []
for i in range(len(trips)):
    no_cols_trips.append(len(trips[i].columns))
no_cols_trips
```

Calcolo il numero minimo e massimo di colonne corrispondenti

```python
print(f'min(no_cols_trips) = {min(no_cols_trips)}')
print(f'max(no_cols_trips) = {max(no_cols_trips)}')
```

Salvo l'indice corrispondente ai dataframe che presentano un numero di colonne minimo

```python
min_cols_idx = []
for i in range(len(cols_trips)):
    if len(cols_trips[i]) == min(no_cols_trips):
        min_cols_idx.append(i)
min_cols_idx
```

```python
for i in range(len(min_cols_idx)):
    print(cols_trips[i])
```

Salvo l'indice corrispondente ai dataframe che presentano un numero di colonne maggiore al numero di colonne minimo

```python
sup_min_cols_trips_idx = []
for i in range(len(cols_trips)):
    if len(cols_trips[i]) > min(no_cols_trips):
        sup_min_cols_trips_idx.append(i)
sup_min_cols_trips_idx
```

L'insieme delle colonne corrispondenti al numero di colonne minimo, cioè l'insieme delle colonne che dovranno contenere tutti i dataframe.

```python
set_min_cols = set(cols_trips[20])
set_min_cols
```

Si calcola l'insieme di colonne da eliminare corrispondenti ai dataframe che presentano un numero di colonne maggiore al numero di colonne minimo.

```python
set_cols_to_delete = []
for i in range(len(cols_trips)):
    curr_set_cols = set(cols_trips[i])
    to_delete = curr_set_cols - set_min_cols
    set_cols_to_delete.append(to_delete)
set_cols_to_delete
```

Quindi, considerando, ad esempio, il primo set di colonne:

```python
set(cols_trips[0])
```

L'insieme di colonne da eliminare è il seguente:

```python
set_cols_to_delete[0]
```

Dal momento che le colonne consetite sono soltanto le seguenti:

```python
set_min_cols
```

Pertanto, si procede a effettuare il drop delle colonne non consetite nei corrispondenti dataframe che presentano un numero di colonne maggiore al numero di colonne minimo

```python
for i in range(len(sup_min_cols_trips_idx)):
    # si prende l'indice corrispondente al dataframe con cols>min_cols
    idx = sup_min_cols_trips_idx[i]
    trips[idx].drop(list(set_cols_to_delete[idx]), axis=1, inplace=True)
    set_cols_to_delete[idx].clear()
```

Considerando, ad esempio, il dataframe 0, si può notare come il set di colonne da eliminare sia vuoto dal momento che tali colonne sono state droppate:

```python
set_cols_to_delete[0]
```

Infatti, considerando sempre il dataframe 0, si può evidenziare come il numero di colonne corrispondente sia pari al numero di colonne minimo:

```python
trips[0]
```

Quindi, si può notare come il numero di colonne per ogni dataframe sia pari al numero di colonne minimo:

```python
new_no_cols_trips = []
for i in range(len(trips)):
    new_no_cols_trips.append(len(trips[i].columns))
new_no_cols_trips
```
