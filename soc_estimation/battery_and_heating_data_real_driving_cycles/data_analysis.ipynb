```python
import os
import pandas as pd
```

```python
data_dir = './data'
measurement_data_dir = f'{data_dir}/measurement'
os.listdir(measurement_data_dir)
```

```python
measurement_data_overview_path = os.path.join(measurement_data_dir, 'Overview.xlsx')
measurement_data_overview = pd.read_excel(measurement_data_overview_path)
measurement_data_overview
```

```python
trips_dir = os.path.join(measurement_data_dir, 'trips')
os.listdir(trips_dir)
```

```python
trips = []
csv_list = [f for f in os.listdir(trips_dir) if f.endswith('.csv')]
for i in range(len(csv_list)):
    trip_name = csv_list[i]
    trip_path = os.path.join(trips_dir, trip_name)
    trip = pd.read_csv(trip_path)
    trips.append(trip)
```
