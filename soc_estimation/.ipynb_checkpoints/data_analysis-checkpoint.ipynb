```python
import os
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

# Definitions

## Paths Declaration
Qui, di seguito, vengono riportati i path relativi al dataset di training, validation e testing. In particolare, il dataset di validation coincide con il primo dataset di testing contenuto nella cartella './data/test/'. Infatti, all'interno della directory './data/test/' sono presenti 5 possibili insiemi di dati così da poter essere sfruttati durante il testing.

```python
data_dir = './data/'
print(f'data_dir = {os.listdir(data_dir)}')

train_data_dir = './data/train/'
print(f'train_data_dir = {os.listdir(train_data_dir)}')

val_data_dir = './data/test/'
print(f'val_data_dir = {os.listdir(val_data_dir)}')

test_data_dir = './data/test/'
print(f'test_data_dir = {os.listdir(test_data_dir)}')
```

# Dataset Analysis and Visualization

```python
train_data_filename = os.listdir(train_data_dir)[0]
train_data_path = os.path.join(train_data_dir, train_data_filename)

train_data = pd.read_csv(train_data_path)
train_data
```

```python
train_data.isna().sum()
```

```python
train_data.info()
```

```python
sns.violinplot(data=train_data)
```

```python
fig, axs = plt.subplots(1, train_data.shape[1], figsize=(20, 4))
for i, key in enumerate(train_data.keys()):
    axs[i].hist(train_data[key], bins=30, alpha=0.75)
    axs[i].legend([f'Feature: {key}'])
plt.tight_layout()
plt.suptitle('Distribution Plots', fontsize=16, y=1.06)
plt.show()
```

```python
fig, axs = plt.subplots(train_data.shape[1], 1, figsize=(14, 6))
for i, key in enumerate(train_data.keys()):
    plt.subplot(train_data.shape[1], 1, i + 1)
    plt.plot(train_data[key])
    plt.legend([f'Feature: {key}'])
plt.tight_layout()
plt.suptitle('Time Series', fontsize=16, y=1.06)
plt.show()
```
